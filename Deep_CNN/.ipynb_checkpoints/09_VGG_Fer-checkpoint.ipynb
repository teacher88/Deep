```python
import numpy as np
import pandas as pd
import seaborn as sns
import random

import matplotlib.pyplot as plt
import matplotlib.pyplot as plt             
import matplotlib as mpl                    
mpl.rc('font', family='Malgun Gothic')      
plt.rcParams['axes.unicode_minus'] = False  

import tensorflow as tf
from tensorflow import keras  
from sklearn import metrics
from keras.models import load_model

from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16

import os
from matplotlib.image import imread
```

- 얼굴 이미지 감성 분류 https://www.kaggle.com/datasets/msambare/fer2013/data
    - angry 화난, disgust 싫음, fer 두려움, happy 행복, neutral 중립적, sad 슬픈, surprise 놀라다
    - 학습 : 28709 * 7
    - 테스트 : 7178 * 7

```python
np.random.seed(3)
tf.random.set_seed(3)
random.seed(42)
```

```python
train_dir="./Data/fer2013/train"
test_dir="./Data/fer2013/test"
```

```python
categories=os.listdir(train_dir)                     #  모든 하위 폴더(감정 카테고리 폴더)를 리스트
print(categories)

image_count={}  
for category in categories:  
    category_path=os.path.join(train_dir, category)  # 각 카테고리 폴더 경로 생성
    print(category_path)
    
    image_list=os.listdir(category_path)             # 해당 폴더 안의 파일 목록 가져오기
    image_count[category]=len(image_list)            # 이미지 개수 저장
print(image_count)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=list(image_count.keys()), 
            y=list(image_count.values()), 
            hue=list(image_count.keys()))
plt.show()
```

```python
plt.figure(figsize=(6, 3))

for i, category in enumerate(categories):
    category_path=os.path.join(train_dir, category)
    image_path=os.path.join(category_path, os.listdir(category_path)[0])  
    # print(image_path)
    
    image=imread(image_path)    
    plt.subplot(2, 4, i+1)
    plt.imshow(image, cmap='gray') 
    plt.title(category)
    plt.axis('off')

plt.tight_layout()   # 레이아웃 자동 조절
plt.show()
```

```python
train_datagen=ImageDataGenerator(rescale=1/255.0,
                                 validation_split=0.2,    # fit()  validation_split=0.2 처리안됨
                                 horizontal_flip=True,    # 뒤집기 (좌우)
                                 width_shift_range=0.1,   # 왼쪽 또는 오른쪽으로 무작위 이동
                                 height_shift_range=0.1,  # 왼쪽 또는 오른쪽으로 무작위 이동
                                 fill_mode='nearest')     # 이동이나 변형으로 인해 생긴 빈 픽셀을 가장 가까운 값으로 채움

train_generator=train_datagen.flow_from_directory(directory=train_dir, 
                                                  target_size=(48, 48),
                                                  batch_size=32,              # 한 번에 가져오는 이미지 개수, 이미지 로드
                                                  class_mode='sparse',       # 또는 categorical (원핫인코딩)
                                                  subset='training')  

validation_generator=train_datagen.flow_from_directory(directory=train_dir, 
                                                      target_size=(48, 48),
                                                      batch_size=32,          # 한 번에 가져오는 이미지 개수, 이미지 로드
                                                      class_mode='sparse',      
                                                      subset='validation')   
train_generator.class_indices
```

```python
test_datagen=ImageDataGenerator(rescale=1/255.0)

test_generator=test_datagen.flow_from_directory(directory=test_dir,
                                                target_size=(48, 48),
                                                batch_size=64,
                                                class_mode='sparse')
test_generator.class_indices
```

```python
# VGG16 신경망 모델
transfer_model=VGG16(weights='imagenet',   
                     include_top=False,    # 완전히 연결된 3개의 계층 (나의 이미지로)
                     input_shape=(48, 48, 3))

transfer_model.trainable=False    # VGG16 모델의 가중치가 학습 중에 업데이트되지 않도록 설정
transfer_model.summary()
```

```python
# VGG16 신경망에 사용자 계층 추가 
model=keras.Sequential()
model.add(transfer_model)

# 완전층
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(64, activation='relu'))

model.add(keras.layers.Dense(7, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0002),
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
history=model.fit(train_generator,
                  epochs=5,        # 50
                  validation_data=validation_generator, 
                  validation_steps=24)      
```

```python
# 평가 
print(model.evaluate(test_generator))
```

```python
# 그래프로 표현
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.plot(history.history['accuracy'], label='Train Accuracy')

plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(history.history['loss'], label='Train Loss')


plt.xlabel('Epoch')
plt.ylabel('Accuracy/Loss')
plt.legend(loc='best')

plt.title('학습 진행에 따른 학습 데이터와 검증 데이터의 정확도/에러')
plt.show()
```

```python
# ImageDataGenerator --> Numpy array

steps=test_generator.n // 5   # 한 번에 가져오는 이미지 개수 (배치 크기)
# print(steps)                  # 24 

images, labels=[],[]

for i in range(steps):
    img, lab=next(iter(test_generator))
    images.extend(img)      # images 리스트에 하나씩 넣기
    labels.extend(lab)

images=np.asarray(images)   # 리스트 → NumPy 배열
labels=np.asarray(labels)
labels[0]
```

## 결과 예측

```python
# 예측 확률
pred_prob=model.predict(images)
```

```python
pred_prob[0]
```

```python
# 결과 예측 
pred=np.argmax(pred_prob, axis=1)
pred[0]
```

```python
fig, axes=plt.subplots(2, 5, figsize=(10, 5))     # 2x5 그리드로 서브플롯 생성
for ax, img, pre, label in zip(axes.flat, images[:10], pred[:10], labels[:10])):
    ax.imshow(img)  # 이미지를 표시
    ax.set_xlabel(f'Pred:{pre}, Real:{label}')    # 라벨을 x축 레이블로 설정

    ax.set_xticks([])     # x축 눈금 제거
    ax.set_yticks([])     # y축 눈금 제거
     
plt.tight_layout()  # 레이아웃 간격 자동 조정
plt.show()
```

```python
# 결과 확인
print(metrics.classification_report(labels, pred))
```
